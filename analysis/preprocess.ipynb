create add_event function to help add the event to the correct location of timestamps.

```python
def add_event(new_df,groundtruth_df):
    start_timestamp_str = new_df['timestamp'].values[0]
    start_timestamp = datetime.strptime(start_timestamp_str, '%d/%m/%Y %H:%M:%S')
    for i in range(0,len(new_df)):
        for j in range(0,len(groundtruth_df)):
            current_date_str = new_df['timestamp'].values[i]
            current_date_obj = datetime.strptime(current_date_str, '%d/%m/%Y %H:%M:%S')
            time_diff = current_date_obj-start_timestamp
            time_diff = time_diff.total_seconds()
#             print(groundtruth_df[' inicio'].values[j])
            if time_diff >=groundtruth_df[' inicio'].values[j] and time_diff <= groundtruth_df[' fim'].values[j]:
                new_df['event'][i] = groundtruth_df['evento'].values[j]      
    return new_df  
```
